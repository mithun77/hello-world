## Bivariate Analysis

* Bivariate analysis is a statistical method that helps you study relationships (correlation) between data sets

### Quantitative vs. Quantitative
    1. Scatterplot
    2. Line plot

### Categorical vs. Quantitative
    1. Bar chart (on summary statistics)
    2. Grouped kernel density plots
    3. Box plots
    4. Violin plots


### Categorical vs. Categorical
    1. Stacked bar chart
    2. Grouped bar chart
    3. Segmented bar chart

### Stacked bar chart

A stacked bar chart, also known as a stacked bar graph, is a graph that is used to break down and compare parts of a whole. Each bar in the chart represents a whole, and segments in the bar represent different parts or categories of that whole. Different colors are used to illustrate the different categories in the bar.



### 1.9a) Create a Stacked bar chart for target & Sex. And write your Observations?

```python
heart[['sex','target']].groupby('sex').mean().plot(kind='bar', stacked=True)
plt.xticks(rotation=320)
plt.show()
```

# 1.10 Quantitative vs. Quantitative

### 1.10a) Create a scatter plot for age & st_depression. And write your Observations?

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x='age', y='oldpeak', data=heart)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.stripplot(x='age', y='oldpeak', data=heart)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.swarmplot(x='age', y='oldpeak', data=heart)
plt.show()
```

It is observed from the above three graphs(scatter, strip, swarm) that the data age and st_depression is not so highly dependent on each other. If we check the correaltion value of them both, we may get very low value. As st_depression is spread across all the age category, we can not derive at a conclusion on their dependency.

With the help of scatter plot, we know their behaviour but redundant datas are not shown in it. It is shown with the help of strip and swarn plot.
