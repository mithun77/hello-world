### 1.1 Import the necessary librairies

### 1.4 Replacing "?" with null values in each column?

### 1.7 Imputing the missing values 

```python
numerical_data = data.select_dtypes(include = np.number)
categorical_data = data.select_dtypes(exclude = np.number)
```

```python
categorical_data.head()
```

```python
numerical_data.isna().sum() # it is evident that numerical_data has no null values
```

```python
categorical_data.isna().sum()
```

Here categorical daata has NULL values, so to treat them, we first need to convert categorical values into numerical values.

```python
categorical_data.info()
```

```python
categorical_data.isna().sum()
```

###  1.8 Imputing null values for categorical variables using get_dummies

```python
categorical_data_get_dummies = pd.get_dummies(categorical_data, drop_first=True)
categorical_data_get_dummies.head()
```

### 1.10 Data Visualization for categorical variables using count plot?

```python
import seaborn as sns


for i in categorical_data.columns:
    sns.countplot(categorical_data[i])
```

### 1.11 Splitting the dataset into dependent and independent variables?

```python
final_data = pd.concat([numerical_data,categorical_data_get_dummies], axis=1)
final_data.head()
```

```python
# out target is Biopsy, hence it is our dependent variable
y=final_data['Biopsy']
X= final_data.drop(columns=['Biopsy'], inplace=True)
```

### 1.12 Splitting the dataset into  training and test set?

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
